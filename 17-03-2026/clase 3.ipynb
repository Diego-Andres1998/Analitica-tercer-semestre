```python
import pandas as pd

# --- PASO 1: CREACIÓN DEL DATASET ---
# Recreamos el DataFrame 'df' basado en la estructura de la imagen
data = {
    'Depto': ['Ventas', 'Ventas', 'Marketing', 'IT', 'Marketing', 'Ventas', None, 'IT'],
    'Monto_CLP': [150000, 150000, 200000, 350000, 200000, 120000, 50000, 350000],
    'Monto_USD': [165.50, 165.50, 220.00, 385.20, 220.00, 132.00, 55.00, 385.20]
}
df = pd.DataFrame(data)

# --- PASO 2: RUTINA DE AUDITORÍA ---
print("--1. TOTAL DE REGISTROS--")
print(f"Filas: {df.shape[0]}")

print("\n--2. DETECCIÓN DE DUPLICADOS--")
print(f"Total Filas Duplicadas: {df.duplicated().sum()}")

print("\n--2.1 DETALLE DE FILAS DUPLICADAS--")
filas_duplicadas = df[df.duplicated(keep=False)]
if not filas_duplicadas.empty:
    print(filas_duplicadas)
else:
    print("No se encontraron filas duplicadas.")

print("\n--3. DETECCIÓN DE NULOS--")
print(df.isnull().sum())

print("\n--4. DETECCIÓN DE INCONSISTENCIA--")
columna_cat = 'Depto'
if columna_cat in df.columns:
    print(f"Valores únicos en '{columna_cat}': {df[columna_cat].unique()}")

# --- PASO 3: TRANSCRIPCIÓN DEL CÓDIGO DE LA IMAGEN ---
print("\n" + "-"*40)
print("--- SCRIPT DE REPORTE AUTOMÁTICO ---")
print("GENERANDO RESUMEN AUTOMÁTICO POR DEPARTAMENTO\n")

departamentos = df['Depto'].dropna().unique()

for depto in departamentos:
    filtro = df[df['Depto'] == depto]
    
    cantidad = len(filtro)
    total_ventas_clp = filtro['Monto_CLP'].sum()
    promedio_usd = filtro['Monto_USD'].mean()
    
    print(f"DEPARTAMENTO: {depto.upper()}")
    print(f" - N° de Transacciones: {cantidad}")
    print(f" - Ingresos Totales: ${total_ventas_clp:,.0f} CLP")
    print(f" - Ticket Promedio: ${promedio_usd:,.2f} USD")
    print("-" * 40)
```

```python
import pandas as pd

# 1. Aseguramos que los datos existan (Dataset del paso anterior)
data = {
    'Depto': ['VENTAS', 'VENTAS', 'MARKETING', 'IT', 'MARKETING', 'VENTAS', 'IT'],
    'Monto_CLP': [150000, 150000, 200000, 350000, 200000, 120000, 350000],
    'Monto_USD': [165.50, 165.50, 220.00, 385.20, 220.00, 132.00, 385.20]
}
df = pd.DataFrame(data)

# 2. Definición de constantes para la auditoría financiera
TASA_IVA = 0.19

print("--- REPORTE FINANCIERO: CÁLCULO DE IVA POR DEPARTAMENTO ---\n")

# 3. Proceso de Transformación Avanzada
departamentos = df['Depto'].dropna().unique()

for depto in departamentos:
    # Filtrado segmentado
    df_depto = df[df['Depto'] == depto]
    
    # Cálculo de métricas solicitadas
    total_bruto_clp = df_depto['Monto_CLP'].sum()
    
    # Cálculo financiero (Bruto = Neto + IVA)
    # Neto = Bruto / (1 + Tasa)
    neto_clp = total_bruto_clp / (1 + TASA_IVA)
    iva_clp = total_bruto_clp - neto_clp
    
    print(f"DEPARTAMENTO: {depto}")
    print(f" > Total Bruto:  ${total_bruto_clp:>12,.0f} CLP")
    print(f" > Total Neto:   ${neto_clp:>12,.0f} CLP")
    print(f" > IVA (19%):    ${iva_clp:>12,.0f} CLP")
    print("-" * 50)

# 4. Resumen Global de Auditoría
total_general = df['Monto_CLP'].sum()
print(f"TOTAL GLOBAL AUDITADO: ${total_general:,.0f} CLP")
```
