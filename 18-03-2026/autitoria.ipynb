```python
import pandas as pd
import numpy as np

# Paso 1: Extracción y Creación del Dataset basado en la imagen proporcionada
data = {
    'ID_Usuario': [1, 2, 2, 3, 4, 5, 6, 7],
    'Nombre': ['Juan Gomez', 'Ana Silva', 'Ana Silva', 'Pedro Perez', 'Luis Jara', 'Marta Diaz', 'Jose Perez', 'Raul V.'],
    'Edad': [25, 30, 30, -5, 150, 28, np.nan, 35],
    'Email': ['juan@gmail.com', 'ana@cl', 'ana@cl', 'pperez@gmail.com', 'l_jara', 'marta@gmail.com', 'jp@gmail.com', 'raul@gmail.com'],
    'Fecha_Registro': ['2024-01-01', '2024-01-01', '2024-01-01', '2024-01-01', '01/01/24', '2024-01-01', '2023-12-01', '2024-01-01']
}

df = pd.DataFrame(data)
```

```python
df.to_csv('autitoria_calidad.csv', index=False)
print("Creado")
```

```python
print("--- REPORTE DE CALIDAD DE DATOS ---")

# # 1. Chequeo de UNICIDAD
duplicados = df.duplicated().sum()
print(f"1. Unicidad: Se encontraron {duplicados} registros duplicados.")

# # 2. Chequeo de COMPLETITUD
nulos = df.isnull().sum().sum()
print(f"2. Completitud: Faltan {nulos} datos en el dataset.")

# # 3. Chequeo de EXACTITUD (Rango de edad lógico: 0 a 120 años)
exactitud = df[(df['Edad'] < 0) | (df['Edad'] > 120)]
print(f"3. Exactitud: Hay {len(exactitud)} registros con edades fuera de rango lógico.")

# # 4. Chequeo de VALIDEZ (Formato de email simple: debe contener '@')
validez = df[~df['Email'].str.contains('@')]
print(f"4. Validez: Hay {len(validez)} correos con formato inválido.")

# # 5. Chequeo de CONSISTENCIA (Formatos de fecha)
# Se identifica 'Nombre' como la columna categórica relevante para la estructura solicitada, 
# aunque el script de la imagen se enfoca en 'Fecha_Registro' para este paso.
print(f"5. Consistencia: Valores únicos en Fecha:\n{df['Fecha_Registro'].unique()}")
```

```python
3. Preguntas de Investigación y Análisis

A. Impacto en el Negocio: Baja Completitud en "Teléfono de Contacto"
Impacto en Actualidad: Si los teléfonos faltan (baja completitud), la empresa 
no puede contactar al cliente en el momento preciso de la campaña. 
Consecuencia: Esto degrada la Actualidad,
 ya que la oportunidad de venta se pierde por no tener el canal de 
 comunicación disponible cuando la oferta aún es vigente. 

B. Diferencia Técnica: Exactitud vs. Validez en la columna "Edad"
Exactitud: Se refiere a si el dato refleja la realidad. En nuestro código, 
una edad de 150 o -5 es inexacta porque no es biológicamente posible, aunque 
el dato sea un número. 
Validez: Se refiere a si el dato cumple con el formato o reglas definidas. 
Si la columna "Edad" espera números enteros y recibe un texto (ej: "treinta"),
 el dato es inválido por tipo de dato. 

C. Herramientas de Python: ¿Por qué .describe() detecta fallas en Exactitud?
Detección de Outliers: Al mostrar el min y max, .describe() permite identificar
 valores que se escapan de la lógica de negocio. 
Ejemplo: En nuestro caso, el valor máximo de 150 y el mínimo de -5 saltarían 
inmediatamente a la vista en un resumen estadístico, señalando una falla de exactitud. 

D. Investigación: Data Governance (Gobernanza de Datos)
Concepto: Es el marco normativo y operativo que define quién tiene el 
control y la autoridad sobre los activos de datos. 
Relación con Dimensiones: La Gobernanza asegura que se cumplan estándares de 
calidad (Exactitud, Completitud, etc.) mediante políticas y procesos auditables, 
evitando que el dataset llegue con errores a la fase de análisis. 

E. Análisis de Caso: Imputación de la edad de Pedro Perez (-5)
Diagnóstico: Si imputamos el valor de -5 con el promedio (digamos 30 años), 
estamos mejorando la Completitud (el campo ya no tiene un valor erróneo o nulo)
 y la Consistencia técnica. 
Justificación: No necesariamente mejoramos la Exactitud, ya que no sabemos 
la edad real de Pedro; solo estamos colocando un valor probable para que el 
modelo estadístico no falle. La exactitud real solo se recupera verificando la fuente original.
```
